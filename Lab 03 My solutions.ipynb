### My solution

```python
# start your code here
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 2000)

y1=0
y2=(-0.9+x)
y3=(4.9-x)

mask1 = (x >= 0.9) & (x <= 2.9)
mask2 = (x >= 2.9) & (x <= 4.9)
mask3 = (x>=4.9)
mask4 = (x<0.9)

y = mask4 * y1 + mask1 * y2 + mask2 * y3 + mask3 * y1


plt.figure()
plt.plot(x,y, label='Piecewise Graph')
plt.grid(True)
plt.legend()
plt.show()
```
